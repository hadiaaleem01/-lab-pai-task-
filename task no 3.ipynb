```python
def water_jug_dfs(c1, c2, target):
    # Stack for DFS: (jug1_current, jug2_current, path_taken)
    stack = [(0, 0, [])]
    visited = set()

    while stack:
        j1, j2, path = stack.pop()

        # Goal Check
        if j1 == target or j2 == target:
            for step in path + [(j1, j2, "Goal Reached!")]:
                print(f"State: ({step[0]}, {step[1]}) - Rule: {step[2]}")
            return True

        if (j1, j2) in visited:
            continue
        visited.add((j1, j2))

        # Define the rules/transitions
        rules = [
            (c1, j2, "Fill Jug 1"),
            (j1, c2, "Fill Jug 2"),
            (0, j2, "Empty Jug 1"),
            (j1, 0, "Empty Jug 2"),
            (j1 - min(j1, c2 - j2), j2 + min(j1, c2 - j2), "Pour Jug 1 -> Jug 2"),
            (j1 + min(j2, c1 - j1), j2 - min(j2, c1 - j1), "Pour Jug 2 -> Jug 1")
        ]

        for next_j1, next_j2, rule_desc in rules:
            if (next_j1, next_j2) not in visited:
                new_path = path + [(j1, j2, rule_desc)]
                stack.append((next_j1, next_j2, new_path))

    print("No solution found.")
    return False

# Parameters: Jug1 Capacity=4, Jug2 Capacity=3, Target=2
water_jug_dfs(4, 3, 2)
```
