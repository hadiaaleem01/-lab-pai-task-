## 1. Reading and Displaying Images


```python
import cv2 
# Read image 
from disk 
img = cv2.imread("geeks.png", cv2.IMREAD_COLOR) 
# Display image 
cv2.imshow("GeeksforGeeks", img) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 2. Saving Images


```python
import cv2 
img = cv2.imread("geeks.png") cv2.imwrite("savedImage.jpg", img) 
# Show saved image img_saved = cv2.imread("savedImage.jpg") 
cv2.imshow("Saved", img_saved) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 3. Resizing Images


```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
image = cv2.imread("geeks.png", 1) half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1) bigger = cv2.resize(image, (1050, 1610)) stretch_near = cv2.resize(image, (780, 540), interpolation = cv2.INTER_NEAREST) Titles = ["Original", "Half", "Bigger", "Interpolation Nearest"] images = [image, half, bigger, stretch_near] 
for i in range(4): plt.subplot(2, 2, i + 1) plt.title(Titles[i]) plt.imshow(images[i]) plt.show()
```

## 4. Color Spaces


```python
import cv2 
src = cv2.imread("geeks.png") 
image = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY) 
cv2.imshow("Grayscale", image) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 5. Arithmetic Operations (Addition & Subtraction)


```python
import cv2 
import numpy as np image1 = cv2.imread("star.jpg") image2 = cv2.imread("dot.jpg") 
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0) 
cv2.imshow("Weighted sum", weightedSum) sub = cv2.subtract(image1, image2) 
cv2.imshow("Subtracted", sub) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 6. Bitwise Operations (AND, OR, XOR, NOT)


```python
import cv2 
import numpy as np img1 = cv2.imread("input1.png") img2 = cv2.imread("input2.png") dest_and = cv2.bitwise_and(img2, img1, mask = None) dest_or = cv2.bitwise_or(img2, img1, mask = None) dest_xor = cv2.bitwise_xor(img1, img2, mask = None) 
cv2.imshow("Bitwise AND", dest_and) 
cv2.imshow("Bitwise OR", dest_or) 
cv2.imshow("Bitwise XOR", dest_xor) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 7. Image Translation


```python
import cv2 
import numpy as np 
image = cv2.imread("geeks.png") height, width = image.shape[:2] quarter_height, quarter_width = height / 4, width / 4 T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]]) img_translation = cv2.warpAffine(image, T, (width, height)) 
cv2.imshow("Translation", img_translation) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 8. Canny Edge Detection


```python
import cv2 
img = cv2.imread("geeks.png") 
edges = cv2.Canny(img, 100, 200) 
cv2.imshow("Edges", edges) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 9. Simple Thresholding


```python
import cv2 
import numpy as np 
image = cv2.imread("geeks.png") 
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
ret, 
thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY) 
cv2.imshow("Binary Threshold", thresh1) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 10. Image Blurring


```python
import cv2 
import numpy as np 
image = cv2.imread("geeks.png") 
Gaussian = cv2.GaussianBlur(image, (7, 7), 0) 
median = cv2.medianBlur(image, 5) 
bilateral = cv2.bilateralFilter(image, 9, 75, 75) 
cv2.imshow("Gaussian Blurring", Gaussian) 
cv2.imshow("Median Blurring", median) 
cv2.imshow("Bilateral", bilateral) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 11. Image Contours


```python
import cv2 
import numpy as np 
image = cv2.imread("geeks.png") 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
edged = cv2.Canny(gray, 30, 200) 
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 
cv2.drawContours(image, 
contours, -1, (0, 255, 0), 3) 
cv2.imshow("Contours", image) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

## 12. Erosion and Dilation


```python
import cv2 
import numpy as np 
img = cv2.imread("geeks.png", 0) kernel = np.ones((5,5), np.uint8) 
img_erosion = cv2.erode(img, kernel, iterations=1) 
img_dilation = cv2.dilate(img, kernel, iterations=1) 
cv2.imshow("Erosion", img_erosion) 
cv2.imshow("Dilation", img_dilation) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
